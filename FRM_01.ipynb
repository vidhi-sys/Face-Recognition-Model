```python
import numpy
import cv2 as cv
from glob import glob
from matplotlib import pyplot as plt
fpath=glob('./data/female/*.jpg')# return images as list
mpath=glob('./data/male/*.jpg')
fpath
```

```python
print('no of images in female', len(fpath))
print('no of images in male' ,len(mpath))
```

```python
img=cv.imread(fpath[0])#read img in bgr format
img_rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
#STEP-02 APPLY HAARCASCADE
haar=cv.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
faces=haar.detectMultiScale(gray,1.5,5)
print(faces)
for x,y,w,h in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    #crop face
    roi=img_rgb[y:y+h,x:x+h]
    plt.imshow(roi)
    plt.axis('off')
    plt.show()
plt.imshow(img)
```

```python
# CROP FEMALE FACES 
for i in range(len(fpath)):
    try:
        
        img = cv.imread(fpath[i])  # read img in bgr format
        img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        # STEP-02 APPLY HAARCASCADE
        haar = cv.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
        gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        faces = haar.detectMultiScale(gray, 1.5, 5)

        for x, y, w, h in faces:
            cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
            # crop face
            roi = img_rgb[y:y + h, x:x + h]
            # SAVE CROP ALL FEMALE FACES 
            cv.imwrite(f'./crop_data/female/female(i).jpg', roi)
            print('image successfully processed')
    except:
        print('cannot process the image')

```

```python
# CROP MALE FACES 
for i in range(len(mpath)):
    try:
        
        img = cv.imread(mpath[i])  # read img in bgr format
        img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        # STEP-02 APPLY HAARCASCADE
        haar = cv.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
        gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        faces = haar.detectMultiScale(gray, 1.5, 5)

        for x, y, w, h in faces:
            cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
            # crop face
            roi = img_rgb[y:y + h, x:x + h]
            # SAVE CROP ALL MALE FACES 
            cv.imwrite(f'./crop_data/male/male(i).jpg', roi)
            print('image successfully processed')
    except:
        print('cannot process the image')
```
